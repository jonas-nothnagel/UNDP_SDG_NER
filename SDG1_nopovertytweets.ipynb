```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_json('UNDP_keywords_SDG1.json', lines = True)
```

```python
df.iloc[12007,:]
```

```python
df = df[['text', 'extended_tweet']]
```

```python
extend_tweet = []
for x in df['extended_tweet']:
    try:
        extend_tweet.append(x['full_text'])
    except:
        extend_tweet.append('')
```

```python
df['extended_tweet'] = extend_tweet
```

```python
tweet_list = []

for x in range(len(df)):
    if len(df['extended_tweet'][x]) != 0:
        tweet_list.append(df['extended_tweet'][x])
    else:
        tweet_list.append(df['text'][x])
        
df['tweet'] = tweet_list
```

```python
df = df[['tweet']]
```
