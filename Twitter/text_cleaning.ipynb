```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('17SDGpt2_tweets.csv', index_col =0,  lineterminator='\n')
```

```python
df
```

```python
import re

def clean_text(df):
    df = df.apply(lambda x: re.sub(r"http\S+", "", x).lower(), 1)\
.apply(lambda i: " ".join(filter(lambda x:x[0]!="@", i.split())), 1)\
.apply(lambda x: re.sub(r"&amp", "",x),1)\
.apply(lambda x: re.sub(r"http\S+", "", x).lower(), 1)\
.apply(lambda x: " ".join(re.sub("[^a-zA-Z]+", " ", x).split()), 1)
    return df
```

```python
df['clean_tweet'] = clean_text(df['tweet'])
```

```python
df.to_csv('cleaned_tweets17SDG.csv')
```
