```python
df = pd.read_csv('17SDGpt2_tweets.csv', lineterminator='\n', index_col = 0)
```

```python
def clean_text(df):
    import re
    df = df.copy().reset_index(drop = True)
    df = df.apply(lambda x: re.sub(r"http\S+", "", x), 1)\
.apply(lambda i: " ".join(filter(lambda x:x[0]!="@", i.split())), 1)\
.apply(lambda x: re.sub(r"&amp", "",x),1)\
.apply(lambda x: re.sub(r"&amp;", "",x),1)
    return df
```

```python
df['clean_tweet'] = clean_text(df['tweet'])
df['clean_tweet'] = df['clean_tweet'].drop_duplicates()
df = df[~df['clean_tweet'].isnull()]
df = df[['sdg', 'clean_tweet']]
```

```python
#Create a class to convert words to tokens 
class tweetDataset(Dataset):
    def __init__(self, tweets, targets, tokenizer, max_len):
        self.tweets = tweets
        self.targets = targets
        self.tokenizer = tokenizer
        self.max_len = max_len
    #length of the tweets that we have     
    def __len__(self):
        return len(self.tweets)
    #takes the index of element from data set
    def __getitem__(self, item):
        #string value of tweet at item
        tweet = str(self.tweets[item])
        #target will be our multiclassifiatio of biz and personal
        target = self.targets[item]   
        
        #the encoder to turn to tokens
        encoding = self.tokenizer.encode_plus(
            #use our tweet 
          tweet,
          add_special_tokens=True,
            #we set this max len to 150
          max_length=self.max_len,
          return_token_type_ids=False,
          pad_to_max_length=True,
          return_attention_mask=True,
          return_tensors='pt',
        )
            #returns a dictionary showing the input id , the attention mask and the targets which is the type and sentiment of tweet
            #converted into a pytorch tensor.  Using type long since we have a classification problem 
        return {
          'tweet_text': tweet,
          'input_ids': encoding['input_ids'].flatten(),
          'attention_mask': encoding['attention_mask'].flatten(),
          'targets': torch.tensor(target, dtype=torch.long)
        }
        
```

```python
#Let us set it to 150 the cut off 
MAX_LEN = 100
BATCH_SIZE = 32
EPOCHS = 6
RANDOM_SEED = 777
```

```python
df_train, df_test = train_test_split(df, test_size=0.4, random_state=RANDOM_SEED)
df_val, df_test = train_test_split(df_test, test_size=0.5, random_state=RANDOM_SEED)
```

```python
#around 8k for training, 1k for test and validation 
df_train.shape, df_val.shape, df_test.shape
```

```python
#wrap into pytorch data loader 
def create_data_loader(df, tokenizer, max_len, batch_size):
    ds = tweetDataset(
        #content of review and taken as numpy 
    tweets=df.clean_tweet.to_numpy(),
        #take the target which is type sentiment 
    targets=df['sdgCode'].to_numpy(),
        #pass in tokenizer
    tokenizer=tokenizer,
        #maximum length 
    max_len=max_len
  )

    return DataLoader( ds, batch_size=batch_size, num_workers=0)
```

```python
#create the loaders 
train_data_loader = create_data_loader(df_train, tokenizer, MAX_LEN, BATCH_SIZE)
val_data_loader = create_data_loader(df_val, tokenizer, MAX_LEN, BATCH_SIZE)
test_data_loader = create_data_loader(df_test, tokenizer, MAX_LEN, BATCH_SIZE)
```

```python
#Building a SDG type classifier 
class SDGClassifier(nn.Module):
    def __init__(self, n_classes):
        super(SDGClassifier, self).__init__()
        self.bert = BertModel.from_pretrained(PRE_TRAINED_MODEL_NAME,return_dict=False)
        self.drop = nn.Dropout(p=0.3)
        self.out = nn.Linear(self.bert.config.hidden_size, n_classes)
        
    def forward(self, input_ids, attention_mask):
        _, pooled_output = self.bert(input_ids=input_ids, attention_mask=attention_mask, return_dict=False)
        output = self.drop(pooled_output)
        return self.out(output)
```

```python
#downoad the pretrained library from hugging Face 
model = SDGClassifier(len(class_names))
model = model.to(device)
```

```python
bert_model = BertModel.from_pretrained(PRE_TRAINED_MODEL_NAME, return_dict=False)
```

```python
#### Training #####
```

```python
optimizer = AdamW(model.parameters(), lr=2e-5, correct_bias=False) 
total_steps = len(train_data_loader) * EPOCHS

scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

loss_fn = nn.CrossEntropyLoss().to(device)
```

```python
def train_epoch(model,data_loader,loss_fn, optimizer,device, scheduler, n_examples):
    model = model.train()
    losses = []
    correct_predictions = 0
    
    for d in data_loader:
        input_ids = d["input_ids"].to(device)
        attention_mask = d["attention_mask"].to(device)
        targets = d["targets"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        
        _, preds = torch.max(outputs, dim=1)
        loss = loss_fn(outputs, targets)
        
        correct_predictions += torch.sum(preds == targets)
        losses.append(loss.item())
        
        loss.backward()

        nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()
    
    return correct_predictions.double() / n_examples, np.mean(losses)
```

```python
#helper function for evaluating the current model 
def eval_model(model, data_loader, loss_fn, device, n_examples):
    model = model.eval()

    losses = []
    correct_predictions = 0

    with torch.no_grad():
        for d in data_loader:
            input_ids = d["input_ids"].to(device)
            attention_mask = d["attention_mask"].to(device)
            targets = d["targets"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            _, preds = torch.max(outputs, dim=1)

            loss = loss_fn(outputs, targets)

            correct_predictions += torch.sum(preds == targets)
            losses.append(loss.item())

    return correct_predictions.double() / n_examples, np.mean(losses)
```

```python
%%time 
#time the total runtime 
#putting the helper functions together 

#to store the training and validation losses, default dictionary that contains a list 
history = defaultdict(list)
#set best accuracy to 0 first and then it will be replaced by best accuracy later 
best_accuracy = 0

#iterate over number of epochs
for epoch in range(EPOCHS):

    #print current epoch
    print(f'Epoch {epoch + 1}/{EPOCHS}')
    #print characters so it's nicer to look at 
    print('-' * 10)

    #return training accuracy and loss of epoch 
    train_acc, train_loss = train_epoch(model, train_data_loader,loss_fn,optimizer,device,scheduler, len(df_train))
    
    #printing the train loss and train accurcy 
    print(f'Train loss {train_loss} accuracy {train_acc}')

    #same thing for the evaluation mode
    val_acc, val_loss = eval_model(model, val_data_loader, loss_fn, device, len(df_val))
    
    #pring the validation loss and accurcy 
    print(f'Val loss {val_loss} accuracy {val_acc}')
    print()
    
    #append the training accuracy, train_loss, validation accurcy, validation loss store into history
    history['train_acc'].append(train_acc)
    history['train_loss'].append(train_loss)
    history['val_acc'].append(val_acc)
    history['val_loss'].append(val_loss)
    
    #if the validation accuracy is higher than the best_accurcy it will replace the number so it originally started at 0
    if val_acc > best_accuracy:
        #saving the best model 
        torch.save(model.state_dict(), 'best_model_state.bin')
        best_accuracy = val_acc
```

```python
#accept a model and a dataloader 
def get_predictions(model, data_loader):
    model = model.eval()

    #4 lists to store data that we need
    tweet_texts = []
    predictions = []
    prediction_probs = []
    real_values = []

    #stuff in evalu model 
    with torch.no_grad():
        for d in data_loader:
            
            #tweets from loader 
            texts = d["tweet_text"]
            input_ids = d["input_ids"].to(device)
            attention_mask = d["attention_mask"].to(device)
            targets = d["targets"].to(device)

            outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask
            )
            _, preds = torch.max(outputs, dim=1)

            probs = F.softmax(outputs, dim=1)
            
            #extend list for text, pred, predictio proability and save targets 
            tweet_texts.extend(texts)
            predictions.extend(preds)
            prediction_probs.extend(probs)
            real_values.extend(targets)
    
    #convert lists into tensors, convert list to single tensor and move to cpu 
    predictions = torch.stack(predictions).cpu()
    #same done for prediction prob and sentiment types 
    prediction_probs = torch.stack(prediction_probs).cpu()
    real_values = torch.stack(real_values).cpu()
    return tweet_texts, predictions, prediction_probs, real_values
```

```python
#get predictions from model and pass in our data 
y_tweet_texts, y_pred, y_pred_probs, y_test = get_predictions(model, test_data_loader)
```

```python
#confusion matrix of predictions 
#helper function for confusion matrix 
def show_confusion_matrix(confusion_matrix):
    hmap = sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Accent_r")
    hmap.yaxis.set_ticklabels(hmap.yaxis.get_ticklabels(), rotation=0, ha='right')
    hmap.xaxis.set_ticklabels(hmap.xaxis.get_ticklabels(), rotation=30, ha='right')
    plt.ylabel('True Tweet Type SDG')
    plt.xlabel('Predicted Tweet Type SDG');
```

```python
#calling our confusion matrix 
cm = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(cm, index=class_names, columns=class_names)
show_confusion_matrix(df_cm)
```
